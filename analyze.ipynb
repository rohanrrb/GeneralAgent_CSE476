```python
import json
from pathlib import Path

test_path = Path("cse_476_final_project_test_data.json")
with test_path.open("r", encoding="utf-8") as fp:
        test_data = json.load(fp)
```

```python
dev_path = Path("cse476_final_project_dev_data.json")
with dev_path.open("r", encoding="utf-8") as fp:
        dev_data = json.load(fp)
```

```python

num = 0
lim = 8192
domain_counts = {}
for Q in dev_data:
    q = Q["input"]
    d = Q["domain"]
    if len(q) > lim:
        print(len(q) - lim)
        print(d)
        num += 1

    if d in domain_counts:
        domain_counts[d] += 1
    else:
        domain_counts[d] = 1

print(num)
print("Counts per domain:", domain_counts)
```
